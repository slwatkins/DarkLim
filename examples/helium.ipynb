```python
# Detector info setup

n_devices = 4
coinc = 2
window = 100e-6 # s

he_gain = 0.15

mass_det = 8.*0.14*1e-3
#mass_det = 0.010 # mass in kg
time_elapsed = 10 # days
efficiency = 1.0

tm = 'He' # target name

energy_res = 0.373e-3 # energy resolution in keV
threshold = coinc*5*energy_res

print('detector mass is {:0.4f} kg = {} g'.format(mass_det,mass_det*1000))
print('run time is {} days'.format(time_elapsed))
print('exposure is {:0.4f} kg-days'.format(mass_det*time_elapsed))
print('threshold = {:0.3f} eV'.format(threshold*1000))
print('assuming {:d}-fold coincidence amongst {:d} devices in a {:0.1f} microsecond window'.format(coinc,n_devices,window/1e-6))

```

```python
def smeared_wimp(en_interp,mass,sigma0=1e-41,tm='He',res=None):
    init_rate = darklim.limit.drde(en_interp, mass, sigma0, tm=tm)
    if res is not None:
        init_rate = darklim.limit.gauss_smear(en_interp, init_rate, res)
    return init_rate
```

```python
# test the converted WIMP spectrum:

fig, ax = plt.subplots(1,figsize=(6,4))

he_gain = 0.15

m_dm = 0.1 # GeV
sig0 = 1e-38 # cm2

eobs = np.geomspace(1e-6,100,1000) # keV, obs energy
edep = darklim.sensitivity.eobs_to_edep(eobs,he_gain) # keV, true dep energy
#dm_func = lambda x: darklim.limit.drde(x, m_dm, sig0, tm=tm)
#obs_dm_func = darklim.sensitivity.drde_obs(eobs, dm_func, he_gain)

dm_func = lambda x: darklim.limit.drde(x, m_dm, sig0, tm=tm)
obs_dm_func = darklim.sensitivity.drde_obs(dm_func, he_gain)


dep_integral = np.trapz(dm_func(edep),x=edep)
obs_integral = np.trapz(obs_dm_func(eobs),x=eobs)
print('deposited energy integral from {:.2e} to {:.2e} keV = {:.2e} evts/kg/day'.format(edep[0],edep[-1],dep_integral))
print('obs energy integral from {:.2e} to {:.2e} keV = {:.2e} evts/kg/day'.format(eobs[0],eobs[-1],obs_integral))
#print('integral from {} to {} keV = {:.2e} evts/day'.format(bins[0],bins[-1],integral))
#print('= {:0.2e} evts/s'.format(integral/24/3600))
#print(' ')

plt.plot(eobs,obs_dm_func(eobs)*np.heaviside(eobs - threshold, 1),label='Observed DM')
plt.plot(edep,dm_func(edep),label='True Deposited DM')

    
ax.set_ylabel('evts/kg/day/keV')
ax.set_xlabel('Energy [keV]')
ax.set_xlim(min(eobs[0],edep[0]),max(eobs[-1],edep[-1]))
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc='lower left',frameon=False)
print('')
#ax.set_ylim(None,1e11)
#ax.legend(loc='upper right',frameon=False,title='{:d} devices total;'.format(n_devices)+' Coinc. level in {:0.0f}'.format(window*1e6)+r'$\mu$s window:',ncol=3)
```

## Reset and run the sens projection

```python
SE = darklim.sensitivity.SensEst(mass_det, time_elapsed, eff=efficiency, tm=tm, gain=he_gain)
SE.reset_sim()

SE.add_flat_bkgd(1) # flat background of 1 DRU
SE.add_nfold_lee_bkgd(m=n_devices,n=coinc,w=window)
```

```python
# run the simulation with FC limits

#m_dms=np.array([0.5])
m_dms=np.geomspace(0.03, 1, num=5),

known_bkgs = [0,1]
e_high = 10 # keV

m_dm, sig, ul = SE.run_sim_fc(
    known_bkgs,
    threshold,
    e_high,
    e_low=threshold,
    m_dms=m_dms,
    nexp=1, # increase for a better estimate, 1 is generally used for diagnostics
    npts=1000,
    plot_bkgd=True,
    res=np.sqrt(n_devices)*energy_res,
    verbose=False,
    sigma0=1e-36,
    use_drdefunction=True
)

print('\nexposure was ',SE.exposure,' kg-days')
print('masses = ',m_dm)
print('ULs (evts) = ',ul)
print('ULs (xsec) = ',sig)
```
